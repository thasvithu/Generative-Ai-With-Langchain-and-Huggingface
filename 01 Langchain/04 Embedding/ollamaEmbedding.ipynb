#### Ollama

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma3:latest")  # default model is 'llama2'
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is a letter in the Greek alphabet.",
        "Beta is the second letter in the Greek alphabet.",
    ]
)

r1
```

```python
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
text = "Alpha is a letter in the Greek alphabet."
r2 = embeddings.embed_query(text)
r2
```
