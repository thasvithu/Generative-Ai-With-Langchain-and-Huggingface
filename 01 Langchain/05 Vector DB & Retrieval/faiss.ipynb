#### FAISS
Facebook AI Similarity Search (FAISS) is a library for efficient similarity search and clustering of dense vectors.
- It is particularly useful for large datasets and high-dimensional vectors.
- FAISS is written in C++ with bindings for Python, and it is optimized for performance on both CPU and GPU.
- FAISS is widely used in various applications, including image and text retrieval, recommendation systems, and natural language processing.
- It provides various indexing structures, such as flat, inverted file, and product quantization, to balance speed and memory usage.
- FAISS also supports approximate nearest neighbor search, which allows for faster retrieval of similar vectors at the cost of some accuracy.
- FAISS is open-source

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader = TextLoader("speech.txt")
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=30)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
db = FAISS.from_documents(docs, embeddings)
```

```python
db
```

```python
query = "What is the main topic of the speech?"
docs = db.similarity_search(query)
docs
```

```python
docs[0]
```

```python
docs[0].page_content
```

```python
query = "How does the speaker describe the desired outcome of the war?"
docs1 = db.similarity_search(query)
docs1
```

##### As a Retriever
we can also convert vectorstore into a Retrive class. THis allows us to easily use it in other LangChain methods, which largely work with retrivers.

```python
retriever = db.as_retriever()
docs = retriever.invoke(query)
docs
```

```python
docs[0].page_content
# The main topic of the speech is the importance of education and its impact on personal and societal growth.
```

#### Similarity Search with score


```python
docs_and_score = db.similarity_search_with_score(query)
docs_and_score
```

```python
embeddings_vector = embeddings.embed_query(query)
embeddings_vector
```

```python
docs_score = db.similarity_search_by_vector(embeddings_vector)
docs_score
```
